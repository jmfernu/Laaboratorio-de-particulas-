## ¿Qué es Docker?
Docker es un proyecto de código abierto que puede implementar automáticamente diferentes aplicaciones en contenedores de software, proporcionando así una capa adicional de abstracción y virtualización de aplicaciones automatizadas en múltiples sistemas operativos. Por ejemplo, cgroups y namespaces, que son maneras por las que Docker consigue optimizar los contenedores en funcionamiento evitando así la sobrecarga y mantener constantes a las máquinas virtuales. Sin lugar a dudas lo interesante de Docker es que puede empaquetar aplicaciones y sus dependencias en contenedores virtuales que pueden ejecutarse en cualquier servidor Linux. Esto ayuda a proporcionar flexibilidad y portabilidad y poder ejecutar los programas en cualquier lugar que haya conexión a internet (obviamente)puede ejecutar la aplicación, ya sea una aplicación local, una nube pública y una nube privada.

## Resume brevemente la historia de Docker¿Por qué Docker se ha convertido en un software tan popular?
La historia de Docker se remonta a un verano de 2010 en donde Kamel Founadi, Solomon Hykes  y Sebastien Pahl idearon y plantaron las bases del primer “Docker” durante el Y Combinator en California.  Aunque verdaderamente la idea ya la traía pensada Solomon Hykes desde Europa donde estuvo trabajando con dotCloud, que era una empresa  asentada en Berlin enfocada en ofrecer a los clientes una manera  de que estos pudieran diseñar y correr sus aplicaciones sin las necesidades y mantenimientos básicos de una aplicación. Docker fue hecho publico en 2013 en la PyCon y en ese mismo año también se hizo “open source” o código abierto para todo el mundo. En 2014 Docker dio un paso muy importante y pasó de usar LXC como ejecutor predeterminando a usar su propio sistema llamado “libcontainer” que estaba escrito en GO, un lenguaje parecido a C pero con características diferentes como la concurrencia –CSP. La fama de Docker ha sido completamente expoenencial. El hecho de que antes no se hubieran hecho contenedores digitales (funcion principla de Docker) era porque no había suficiente “know-how”  y no fue así hasta que los mayores veinte informaticos especializados en Linux decidieron inciar esta aventura. Su mayor fama ha sido sin duda cuando en 2016 Microsoft anunció que le daba soporte en Windows 10 y Windows Server 2016. 

Docker es tan famoso ya que  gracias a su innovador sistema de contenedores ha revolucionado el software . Los creadores de Docker han conseguido hacer sencillo crear contenedores virtuales, han simplificado la manera de hacerlos gracias a la forma de hacer los OS syscalls o llamadas al sistema, que son la forma por la que una app se conecta al núcleo de del sistema. Realmente si nos ponemos críticos podemos diferenciar alrededor de  cuatro “why” de su fama, empezaremos por el primero: 

1.La facilidad de usarlo: 
Al ser open-source y poder ser adaptado en Windows/Mac o en su propio lenguaje Linux lo hace accesible a todo el que quiere trastear con él.  

2.Escalados de sistemas más rápido: con esto nos referimos  a que lo que antes de que Docker fuera inventado, cada cliente necesitaba si quería por ejemplo optimizar su applicaicon (por ejemplo) comprar nuevo hardware lo que resultaba mucho costo al final. Mientras que si con Docker es la propia empresa (Docker) la que ofrece el servicio para el resto deempresas, estas se ahorran un dinero.  

3.Mejor entrega de software: al ser un programa el cual se auto almacena , lo cual lo hacer más fácil de compartir y con ello más eficiente.  

4.Flexibilidad: las aplicaciones que están integradas en contenedores tienen un mejor rendimiento que las que no lo están. 



## ¿Qué es un fichero ppk?¿Qué es una clave RSA público/privada(fichero key)?
Un fichero ppk es un tipo de formato de archivo,  su objetivo es la seguridad de autentificación en la utilización del propio archivo. Poniendo un ejemplo; una parte A y otra parte B tienen que tener ambos una clave por la cual son los unicos que pueden acceder a ese fichero como el algoritmo de Diffie-Hellman.  

RSA en sí es un algoritmo de criptografía asimétrico de clave pública, que combina una clave privada (de cada puerto o destino y unica para cada uno) y una clave pública (accesible para todo el mundo).  Cada clave tiene una utilización distinta, la clave publica es la primera que tiene el turno de salida ya que antes de que un usuario haga mandar los datos a un servidor los cifra con esa clave publica. En segundo lugar la clave privada será la encargada de descifrar los datos hechos por la clave publica. Este tipo de criptografía se ve por ejemplo en el algoritmo de Diffle-Hellman el cual guarda relación en las claves públicas y privadas

## ¿Qué es un positrón?¿Quién postuló su existencia teórica y cuándo y cómo se descubrió experimentalmente?
Un positrón o un antielectrón es un tipo de antimateria  es un tipo básico de partícula (porque no hay evidencia de que esté compuesto por otras partículas más simples) y su carga es igual a la de un electrón, aunque es una carga positiva. Debido a esta característica, el positrón se denomina antipartícula de la partícula subatómica. Los positrones son la  contrapartida de las partículas, sus cargas eléctricas están invertidas pero sus masas son idénticas.  

A finales de la década de 1920, el físico británico Paul Dirac creía que existían los positrones. Finalmente, en 1932, el estadounidense Carl David Anderson (Carl David Anderson) logró demostrarlo estudiando los rayos cósmicos. Los rayos cósmicos son partículas que caen en la Tierra desde el espacio, en general suelen ser protones o núcleos de átomos. Estas partículas viajan a la velocidad de la luz y tienen más potencia que las partículas producidas por el ser humano en el acelerador de partículas más potente del mundo. Ha sido la primera antipartícula que se descubrió la cual la hace todo un hito de la física.  Sobretodo hay que remarcar  cómo Carl David la descubrió, que fue gracias a una cámara de niebla o de Wilson también llamada, por la que consiguió diferencias unas partículas que estaban cargadas positivamente y eran mucho mas pequeños que los electrones. Por lo que decidió continuar investigando y llegando a la conclusión de que era un antielectrón.
